```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(data=tips, x='sex',y='total_bill',estimator=np.std)#we plot on categorical values
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')#those dots are outliers
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)#very similar to boxplot and it contains all data points
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)#dodge is like split
```

```python
#swarmplot is combo of stripplot and violinplot
#not recommended for large datasets
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#we can combine violinplot and swarmplot for better understanding
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
