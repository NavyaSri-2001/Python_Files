```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
p=sns.PairGrid(iris)
p
```

```python
p.map(plt.scatter)
plt.show()
```

```python
p.map_diag(plt.hist)
p.map_offdiag(plt.scatter)
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.kdeplot,'total_bill')
```

```python
p=sns.FacetGrid(data=tips,row='smoker',col='time')
p.map(plt.scatter,'total_bill','tip')
```
