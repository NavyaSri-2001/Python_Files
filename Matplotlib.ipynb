```python
import matplotlib.pyplot as plt
#below line allows us to see plots inside this jupiter notebook
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)#args= no.of rows, no.of columns, plot number
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
#Object Oriented, here we make figure as an object and call methods from it
#create figure (empty canvas)
fig=plt.figure()
# Add set of axes to figure
axes= fig.add_axes([0.1,0.1,0.8,0.8])#left,bottom,width,height - range[0,1]
# Plot on that set of axes
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('SET TITLE')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('LARGE PLOT')

axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x,'r')
axes2.set_title('SMALL PLOT')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) #by using subplots() it directly automates add_axes() function based on no.of rows and cols
#axes.plot(x,y)
#axes is just a list, which means we can iterate through it
# Axes is an array of axes to plot on
for current_ax in axes:
  current_ax.plot(x,y)
  current_ax.set_xlabel('X LABEL')
  current_ax.set_ylabel('Y LABEL')
  current_ax.set_title('TITLE')

plt.tight_layout() #this takes care of overlapping issues
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('PLOT 1')

axes[1].plot(y,x)
axes[1].set_title('PLOT 2')
```

**Figure Size and DPI**

```python
fig=plt.figure(figsize=(8,2),dpi=1000)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)
axes[0].plot(x,y,'r')
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#Saving a figure
fig.savefig('my_picture.png')
```

**Adding Legend**

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**3)
axes.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label='X sqaured')
axes.plot(x,x**3, label='X cubed')
axes.legend(loc=0)#0 is best location code, it puts legend at best location
```

**PLOT LOOK**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, color='yellow',linewidth=3,alpha=0.3, linestyle='-')#google RGB Hex code for different color code, alpha indicates amount of transparency
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=5,ls='-.')
```
