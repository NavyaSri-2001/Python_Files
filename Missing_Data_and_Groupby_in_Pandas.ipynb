```python
#Missing Data
import pandas as pd
import numpy as np
```

```python
d={'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
df=pd.DataFrame(d)
df
```

```python
#dropping missing values
df.dropna()#by default it removes rows which have null values
```

```python
df.dropna(axis=1)#to drop columns with null values
```

```python
df.dropna(thresh=2)#thresh means if we have atleast 2 non null values in a row, then dont drop it
```

```python
df.dropna(axis=1, thresh=2)
```

```python
#replacing missing values
df.fillna(value='Value')
```

```python
df['A'].fillna(value=df['A'].mean())#we usually fill null values with mean of that column
```

```python
#Group By
d={'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],'Name':['Sam','Jack','Lara','Ash','Pari','Shiv'],'Sales':[100,200,300,400,500,600]}
df=pd.DataFrame(d)
df
```

```python
byComp= df.groupby('Company')
byComp
```

```python
byComp.sum()
```

```python
byComp.sum().loc['FB']
```

```python
#doing in a single line
df.groupby('Company').sum().loc['FB']
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```
