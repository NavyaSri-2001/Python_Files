```python
import pandas as pd
```

```python
df1=pd.read_csv("/content/LOTR.csv")
df1
```

```python
df2=pd.read_csv("/content/LOTR 2.csv")
df2
```

**Default will be inner join**

```python
df1.merge(df2)
```

**Though firstName is same in both tables, since we are not joining on that we are getting 2 seperate columns**

```python
df1.merge(df2, how='inner', on='FellowshipID')
```

```python
df1.merge(df2, how='inner', on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,how='cross')
```

**Joins**

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_left', rsuffix='_right')
```

**Joins work better with indexes**

```python
df4= df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how='outer', lsuffix='_left',rsuffix='_right')
df4
```

**Merge is more easier compared with joins.
Merge joins based on columns,
joins join based on indexes**

**Concatenate**
**It is like union in sql. It joins data frames one below other unlike merge, join which join dfs side by side**

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='inner')
```

```python
pd.concat([df1,df2],axis=1)
```
