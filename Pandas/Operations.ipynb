```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'col1':[1,2,3,4],'col2':[444,555,666,444],'col3':['abe','def','ghi','xyz']})
df.head()
```

```python
df['col2'].unique()#finding unique values
```

```python
#finding no.of unique values in 2 ways
print(len(df['col2'].unique()))
print(df['col2'].nunique())
```

```python
df['col2'].value_counts()#counting no.of occurances of each element
```

```python
#conditional selections
df[(df['col2']>2) & (df['col2']==444)]
```

```python
df[['col1','col2']].sum(axis=1)
```

```python
#applied method
def times2(x):
  return x*2
```

```python
df['col1'].apply(times2)
```

```python
#we can also pass built in functions in apply
df['col3'].apply(len)
```

```python
df['col2'].apply(lambda x: x*2)#apply is more useful with lambda functions
```
