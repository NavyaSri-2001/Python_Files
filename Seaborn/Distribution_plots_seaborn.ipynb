```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips') #seaborn contains some built-in datasets, one of them is tips
tips.head()
```

```python
sns.distplot(tips['total_bill'])#this takes only 1 variable
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)#this is a bi-variable plot
```

```python
sns.pairplot(tips)#it plots pairs of all numerical columns
#for same row vs col we get histogram
#hue takes only categorical values
```

```python
#hue takes only categorical values
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])#it is similar to distplot. We get dashes instead of bars
```

```python
sns.kdeplot(tips['total_bill'])#same as ditplot without bars
```
