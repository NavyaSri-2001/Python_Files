**Data Frames**

```python
from numpy.random import randn
np.random.seed(101)#seed gives same numbers every time
df=pd.DataFrame(randn(5,4),['A','B','C','D','E'],['P','Q','R','S'])
df
```

Above each P, Q, R, S is a series
A dataframe is a buch of series with common indexes

```python
df['P']
```

```python
type(df['P'])
```

```python
type(df)
```

```python
df.P #we can access a column like this also, but not frequently used
```

```python
df[['P','Q']]
```

```python
df['new']=df['P']+df['Q']#creating a new column or series
df
```

```python
#droping or deleting a column
df.drop('new',axis=1)
```

```python
df
```

```python
df.drop('new',axis=1,inplace=True)#to save the change
df
```

```python
df.drop('E',inplace=True)#deleting row, axis=0 is not required becoz axis=0 is by defualt set
df
```

```python
#accessing rows
df.loc['C']#loc stands for location
```

```python
df.iloc[2]#iloc - integer location, it takes int as input
```

```python
df.loc['A','P']
```

```python
#accessing subset of dataframe
df.loc[['A','B'],['P','R']]
```

```python
#conditional selection
booldf=df>0
booldf
```

```python
df[booldf]#gives only values which have True
```

```python
df[df>0]
```

```python
df['P']>0
```

```python
df[df['P']>0]#gives all rows which dont have null values
```

```python
df[df['R']>0]
```

```python
result_df=df[df['R']>0]
result_df
```

```python
result_df['S']#since result_df is also a dataframe, we can extract whichever data we want
```

```python
#doing above steps in a single line
df[df['R']>0][['P','Q']]
```

```python
#multiple conditionals
df[(df['R']>0) & (df['S']>0)]#we can't use "and" keyword here, it gives error
```

```python
df[(df['R']>0) | (df['S']>0)]
```

```python
#resetting index to original numerical
df.reset_index(inplace=True)
df
```

```python
df.drop('index',axis=1, inplace=True)
```

```python
df
```

```python
#creating a new index and setting it
newindex= 'CA NY AB MN, KR'.split() #split() method splits a string into list of strings based on white space
newindex
```

```python
df['States']=newindex
df
```

```python
df.set_index('States', inplace= True)
df
```

```python
#Multi indexing and hier index
#Index levels
outside=['G1','G1','G1','G2','G2','G2']
inside=[1,2,3,1,2,3]
hier_index=list(zip(outside,inside))
hier_index=pd.MultiIndex.from_tuples(hier_index)
```

```python
outside
```

```python
list(zip(outside,inside))
```

```python
hier_index
```

```python
df=pd.DataFrame(randn(6,2),hier_index,['A','B'])
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[2]
```

```python
df.index.names=['Groups','Num']
df
```

```python
df.loc['G2'].loc[3]['B']#selecting a value using indexing
```
