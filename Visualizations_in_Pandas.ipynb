```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Ice Cream Ratings.csv")
df= df.set_index('Date')
df
```

```python
print(plt.style.available)
#plt.style.use('dark_background')
plt.style.use('classic')
```

```python
df.plot()
```

```python
df.plot(kind='line', subplots=True)
```

```python
df.plot(kind='line', xlabel='Daily Ratings', ylabel='Scores', title='Ice cream Ratings')
```

```python
df.plot(kind='bar')
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```
